## Scrape the NBC New York homepage

I want a CSV file called `nbc.csv` that includes a row for each story, with these two columns:

* `title`, the title of the story
* `url`, the full URL to the story

If you want to start by printing these out, that's fine, but the end result is hopefully a CSV. Hot tip: use `.strip()` to remove all of the tabs from the left and right of your titles.

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.nbcnewyork.com/")
doc = BeautifulSoup(response.text)
```

```python
stories = doc.find_all(class_='story-card__title-link')
for story in stories:
    print(story['href'])
    print(story.text.strip())
```

```python
nbc_scrape = []

stories = doc.find_all(class_='story-card__title-link')
for story in stories:
    nbc_data = {}
    nbc_data['link'] = story['href']
    nbc_data['title'] = story.text.strip()
    nbc_scrape.append(nbc_data)
nbc_scrape
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(nbc_scrape)
df
```

```python
df.to_csv("nbc.csv")
```
